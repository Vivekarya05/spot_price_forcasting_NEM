```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

plt.style.use("seaborn-v0_8")
sns.set_context("talk")

# Create folder for images
os.makedirs("outputs/figures", exist_ok=True)

# Load dataset
df = pd.read_csv(
    "model_features_dataset.csv",
    parse_dates=["timestamp"]
)

df.head()

```

## Spot Price Time Series (Historical) 

```python
plt.figure(figsize=(14,5))
plt.plot(df["timestamp"], df["target_rrp"], linewidth=1)
plt.title("Spot Price Time Series")
plt.ylabel("Price ($/MWh)")
plt.xlabel("Time")
plt.tight_layout()
plt.savefig("outputs/figures/01_spot_price_timeseries.png")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["target_rrp"], bins=100, kde=True)
plt.xlabel("Spot Price ($/MWh)")
plt.title("Distribution of Spot Prices")
plt.tight_layout()
plt.savefig("outputs/figures/02_spot_price_distribution.png")
plt.show()
```

```python
sample = df.sample(4000, random_state=42)

plt.figure(figsize=(8,6))
plt.scatter(sample["net_demand_after_pv"], sample["target_rrp"], alpha=0.3)
plt.xlabel("Net Demand After PV (MW)")
plt.ylabel("Spot Price ($/MWh)")
plt.title("Net Demand vs Spot Price")
plt.tight_layout()
plt.savefig("outputs/figures/03_net_demand_vs_price.png")
plt.show()
```

```python
hourly_price = df.groupby("hour")["target_rrp"].mean()

plt.figure(figsize=(8,5))
hourly_price.plot(marker="o")
plt.xlabel("Hour of Day")
plt.ylabel("Average Spot Price ($/MWh)")
plt.title("Average Spot Price by Hour of Day")
plt.grid(True)
plt.tight_layout()
plt.savefig("outputs/figures/04_price_by_hour.png")
plt.show()
```

```python
sample = df.sample(4000, random_state=42)

plt.figure(figsize=(8,6))
plt.scatter(sample["pv_rooftop_mw"], sample["net_demand_after_pv"], alpha=0.3)
plt.xlabel("Rooftop PV Generation (MW)")
plt.ylabel("Net Demand After PV (MW)")
plt.title("Solar PV Shaping Effect on Net Demand")
plt.tight_layout()
plt.savefig("outputs/figures/05_pv_vs_net_demand.png")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df["TOTALINTERMITTENTGENERATION"], df["target_rrp"], alpha=0.3)
plt.xlabel("Total Intermittent Generation (MW)")
plt.ylabel("Spot Price ($/MWh)")
plt.title("Intermittent Generation vs Spot Price")
plt.tight_layout()
plt.savefig("outputs/figures/06_intermittent_gen_vs_price.png")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df["temperature"], df["TOTALDEMAND"], alpha=0.3)
plt.xlabel("Temperature (Â°C)")
plt.ylabel("Total Demand (MW)")
plt.title("Temperature vs Electricity Demand")
plt.tight_layout()
plt.savefig("outputs/figures/07_temperature_vs_demand.png")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df["abs_net_demand_ramp"], df["target_rrp"], alpha=0.3)
plt.xlabel("Absolute Net Demand Ramp (MW)")
plt.ylabel("Spot Price ($/MWh)")
plt.title("Demand Ramps vs Spot Price")
plt.tight_layout()
plt.savefig("outputs/figures/08_demand_ramp_vs_price.png")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df["abs_renewable_ramp"], df["target_rrp"], alpha=0.3)
plt.xlabel("Absolute Renewable Ramp (MW)")
plt.ylabel("Spot Price ($/MWh)")
plt.title("Renewable Ramps vs Spot Price")
plt.tight_layout()
plt.savefig("outputs/figures/09_renewable_ramp_vs_price.png")
plt.show()
```

```python
hourly_net_demand = df.groupby("hour")["net_demand_after_pv"].mean()

plt.figure(figsize=(8,5))
hourly_net_demand.plot(marker="o", color="green")
plt.xlabel("Hour of Day")
plt.ylabel("Average Net Demand (MW)")
plt.title("Average Daily Net Demand Profile")
plt.grid(True)
plt.tight_layout()
plt.savefig("outputs/figures/10_net_demand_daily_profile.png")
plt.show()

```

```python
price_kpis = {
    "Average Price ($/MWh)": df["target_rrp"].mean(),
    "Median Price ($/MWh)": df["target_rrp"].median(),
    "Max Price ($/MWh)": df["target_rrp"].max(),
    "Min Price ($/MWh)": df["target_rrp"].min(),
    "95th Percentile Price ($/MWh)": df["target_rrp"].quantile(0.95)
}

price_kpi_df = pd.DataFrame(price_kpis, index=["Value"]).T
price_kpi_df
```

```python
risk_kpis = {
    "Price Volatility (Std Dev)": df["target_rrp"].std(),
    "% Intervals > $300": (df["target_rrp"] > 300).mean() * 100,
    "% Intervals < $0": (df["target_rrp"] < 0).mean() * 100,
    "Max Absolute Price Ramp": df["target_rrp"].diff().abs().max()
}

risk_kpi_df = pd.DataFrame(risk_kpis, index=["Value"]).T
risk_kpi_df
```

```python
time_kpis = {
    "Highest Avg Price Hour": df.groupby("hour")["target_rrp"].mean().idxmax(),
    "Highest Avg Price Day (0=Mon)": df.groupby("dayofweek")["target_rrp"].mean().idxmax(),
    "Highest Avg Price Month": df.groupby("month")["target_rrp"].mean().idxmax()
}

time_kpi_df = pd.DataFrame(time_kpis, index=["Value"]).T
time_kpi_df

```

```python
high_solar = df["pv_rooftop_mw"] > df["pv_rooftop_mw"].quantile(0.75)
low_solar  = df["pv_rooftop_mw"] < df["pv_rooftop_mw"].quantile(0.25)

renewable_kpis = {
    "Avg Price (High Solar)": df.loc[high_solar, "target_rrp"].mean(),
    "Avg Price (Low Solar)": df.loc[low_solar, "target_rrp"].mean(),
    "Solar Price Reduction ($/MWh)": 
        df.loc[low_solar, "target_rrp"].mean() - df.loc[high_solar, "target_rrp"].mean()
}

renewable_kpi_df = pd.DataFrame(renewable_kpis, index=["Value"]).T
renewable_kpi_df
```

```python
high_net_demand = df["net_demand_after_pv"] > df["net_demand_after_pv"].quantile(0.75)
low_net_demand  = df["net_demand_after_pv"] < df["net_demand_after_pv"].quantile(0.25)

demand_kpis = {
    "Avg Price (High Net Demand)": df.loc[high_net_demand, "target_rrp"].mean(),
    "Avg Price (Low Net Demand)": df.loc[low_net_demand, "target_rrp"].mean(),
    "Demand Price Uplift ($/MWh)": 
        df.loc[high_net_demand, "target_rrp"].mean() - df.loc[low_net_demand, "target_rrp"].mean()
}

demand_kpi_df = pd.DataFrame(demand_kpis, index=["Value"]).T
demand_kpi_df
```

```python
plt.figure(figsize=(10,5))
plt.bar(price_kpi_df.index, price_kpi_df["Value"])
plt.xticks(rotation=45)
plt.title("Key Spot Price KPIs")
plt.ylabel("Value")
plt.tight_layout()
plt.savefig("outputs/figures/kpi_price_summary.png")
plt.show()
```
